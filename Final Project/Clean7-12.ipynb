```python
def json_to_df(rootdir):
    
    df_list = []
    artists = []
    tags = []
    track_ids = []
    titles = []
    columns = ['titles', 'artists', 'tags', 'track_ids']
    
    for subdir, dirs, files in os.walk(rootdir):
        for file in files:
            if file.endswith('.json'):
                f = open(os.path.join(subdir, file), 'r')

                data = json.load(f)

                artists.append(data['artist'])

                tag_names = []
                for i in range(0, len(data['tags'])):
                    tag_names.append(data['tags'][i][0])

                tags.append(tag_names)

                track_ids.append(data['track_id'])

                titles.append(data['title'])
    
    df = pd.DataFrame(list(zip(titles, artists, tags, track_ids)), columns = columns)
    
    return df
```
